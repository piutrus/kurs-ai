```python
def arithmetic_mean(list):
  _sum = 0
  _len = 0
  for i in list:
    _sum += i
    _len += 1
  return _sum/_len

def median(list):
  list.sort()
  if len(list)%2 == 0:
    return(list[int(len(list)/2 - 1)])
  else:
    return((list[int(len(list)/2 - 0.5)] + list[int(len(list)/2 + 0.5)])/2)

#def mode(list):
#  _appearances = {}
#  for i in list:
#    if i in _appearances:
#      _appearances[i] += 1
#    else:
#      _appearances[i] = 1

def mode(list_of_numbers):
  count_occurrence = {i: list_of_numbers.count(i) for i in set(list_of_numbers)}
  mode = [key for value, key in zip(count_occurrence.values(),
  count_occurrence.keys()) if value == max(count_occurrence.values())]
  return mode

def variance(list):
  _mean = arithmetic_mean(list)
  _diffs = [(list[i] - _mean)**2 for i in range(len(list))]
  return sum(_diffs)/(len(_diffs)-1)



list1 = [2,3,9,-6,4]

print(arithmetic_mean(list1))
print(median(list1))
print(mode(list1))
print(variance(list1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(69)

X = np.linspace(0, 50, 100)
noise = np.random.randint(-5, 5, 100)

Y = 2*X - 10 + noise

fig, ax = plt.subplots(figsize=(8,5))

ax.scatter(X, Y, alpha=.6)

a = (100*sum(X*Y) - sum(X)*sum(Y))/(100*sum(X**2) - sum(X)**2)
b = (sum(Y) - a*sum(X))/100
a1 = round(a, 2)
b1 = round(b, 2)

ax.plot(X, a*X + b, label=r'$y \approx {}x + {}$'.format(a1, b1), c='r', ls='--')

plt.legend()
plt.grid()
plt.plot()
```
