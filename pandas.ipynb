```python
import pandas as pd
import numpy as np

data = pd.read_excel('Gigathon_Wyniki.xlsx')

df = pd.DataFrame(data)

df
```

```python
print(df.head(5))
print(40*'-')
print(df.tail(5))
```

```python
df['Wynik procentowy'] = round(df['Punktacja']/120 * 100, 2)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.sort_values(by='Wynik procentowy', ascending=False)
```

```python
df['Zakwalifikowany'] = df['Wynik procentowy'] > 70
df
```

```python
df.groupby(['Zakwalifikowany', 'Wiek']) [['Punktacja', 'Wynik procentowy']].agg(['count', 'min', 'max', 'mean', 'median'])
```

```python
df_zakwalifik = df[df['Zakwalifikowany']]
df_zakwalifik.drop('Zakwalifikowany', axis=1, inplace=True)
df_zakwalifik.to_excel('Zakwalifikowani.xlsx')
df_zakwalifik
```

```python
df_niezakwalifik = df[~df['Zakwalifikowany']]
df_niezakwalifik.drop('Zakwalifikowany', axis=1, inplace=True)
df_niezakwalifik.to_excel('Niezakwalifikowani.xlsx')
df_niezakwalifik
```
